# Cleanup Merino Wool Sweater HTML

This notebook reads the HTML file and removes duplicate copies of the document, leaving one clean version.

```python
# Load and clean the HTML file
import pathlib

file_path = pathlib.Path('products/merino-wool-sweater.html')
text = file_path.read_text()
first_idx = text.find('<!DOCTYPE html>')
second_idx = text.find('<!DOCTYPE html>', first_idx + 1)

print('first idx', first_idx, 'second idx', second_idx)

if second_idx != -1:
    new_text = text[second_idx:]
    file_path.write_text(new_text)
    print('Rewrote file, new length', len(new_text))
else:
    print('Only one copy; no change.')
```
